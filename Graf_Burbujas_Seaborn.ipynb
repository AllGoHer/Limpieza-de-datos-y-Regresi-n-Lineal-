```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

peliculas = pd.read_csv("C:\\Users\\ALLAN\\Desktop\\Business Analitics con Python y ChatGPT\\peliculas.csv")
peliculas = peliculas.dropna()
peliculas['Estreno'] = pd.to_datetime(peliculas['Estreno'])

cols_monetarias = ['Presupuesto', 'TaquillaInternacional', 'TaquillaEEUU']

for col in cols_monetarias:

    peliculas[col] = peliculas[col].replace('[\$,]', '', regex=True).astype('int64')

peliculas.head()

peliculas['Ao'] = peliculas['Estreno'].dt.year 
plt.figure(figsize=(10,5), dpi=300)

grafico = sns.scatterplot(x='Ao',

                          y='Presupuesto',

                          size='TaquillaInternacional',

                          hue='TaquillaEEUU',

                          data=peliculas)

plt.show()
```
