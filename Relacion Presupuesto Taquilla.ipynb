```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('C:\\Users\\ALLAN\\Desktop\\Business Analitics con Python y ChatGPT\\peliculas.csv')
```

```python
df['Presupuesto'] = df['Presupuesto'].replace('[\$,]', '', regex=True).astype('int64')

df['TaquillaEEUU'] = df['TaquillaEEUU'].replace('[\$,]', '', regex=True).astype('int64')

df['TaquillaInternacional'] = df['TaquillaInternacional'].replace('[\$,]', '', regex=True).astype('int64')
```

```python
df['Presupuesto'] /= 1e6

df['TaquillaEEUU'] /= 1e6

df['TaquillaInternacional'] /= 1e6
```

```python
# Configurar el estilo y tamao de la figura

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5), dpi=300)



# Crear el primer grfico de regresin para Taquilla en EE.UU.

plt.subplot(1, 2, 1)

sns.regplot(x='Presupuesto', y='TaquillaEEUU', data=df, color='red')

plt.title('Relacin entre Presupuesto y Taquilla en EE.UU.')

plt.xlabel('Presupuesto en Millones de USD')

plt.ylabel('Taquilla en Millones de USD')



# Crear el segundo grfico de regresin para Taquilla Internacional

plt.subplot(1, 2, 2)

sns.regplot(x='Presupuesto', y='TaquillaInternacional', data=df, color='blue')

plt.title('Relacin entre Presupuesto y Taquilla Internacional')

plt.xlabel('Presupuesto en Millones de USD')

plt.ylabel('Taquilla en Millones de USD')



# Mostrar los grficos

plt.tight_layout() # Ajuste automtico de dimensiones del grfico

plt.show()

# Configurar el estilo y tamao de la figura

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5), dpi = 300)



# Crear el grfico de regresin para Taquilla en EE.UU.

sns.regplot(x='Presupuesto', y='TaquillaEEUU', data=df, color='red', scatter_kws={'alpha': 0.1})

plt.title('Relacin entre Presupuesto y Taquilla en EE.UU. y Taquilla Internacional')

plt.xlabel('Presupuesto en Millones de USD')

plt.ylabel('Taquilla en Millones de USD')



# Superponer el grfico de regresin para Taquilla Internacional

sns.regplot(x='Presupuesto', y='TaquillaInternacional', data=df, color='blue', scatter_kws={'alpha': 0.1})



# Mostrar el grfico

plt.show() 
```
